### 3. Prepare Data

**Data preparation**

*Missing Values*
* beer_abv - mean value
* brewery_name - Mode Value - 'Boston Beer Company (Samuel Adams)'
* review_profilename - Mode Value - 'northyorksammy'

*Label Encoding*
* beer_style_cat - Target - Label encode beer_style

*Scaling*
* Use MinMax Scaling for 'beer_abv','review_aroma','review_appearance','review_palate','review_taste'
*Reference Data*
* Beer Style - Used to map label beer_style_cat to beer_style text
* Breweries - Used to map label brewery_id to brewery_name (and vice versa)

```python
# Task: Create a copy of df and save it into a variable called df_cleaned
df_cleaned = df.copy()
```

```python
# Numerical Fields - fill beer_abv with mean value for column
df_cleaned['beer_abv'].fillna(df_cleaned['beer_abv'].mean(), inplace=True)
```

```python
# Categorical Field - brewery_name - fill beer_abv with mode value for column - Boston Beer Company (Samuel Adams)
df_cleaned['brewery_name'] = df_cleaned['brewery_name'].fillna('Boston Beer Company (Samuel Adams)')
```

```python
df_cleaned[df_cleaned['brewery_name'].isna()]
```

```python
# Task: Find mode for review_profilename column
df_review_profilename_mode = df_cleaned['review_profilename'].mode()
print(df_review_profilename_mode)
```

```python
# Categorical Field - review_profilename - fill beer_abv with mode value for column - northyorksammy
df_cleaned['review_profilename'] = df_cleaned['review_profilename'].fillna('northyorksammy')
```
